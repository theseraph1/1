# Equations horaires

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0 = 50
i = np.pi/4
g = 9.81
```

```python
dt = 0.2
t = np.linspace(0,7.25,30)
```

```python
x = v0*np.cos(i)*t
y = -0.5*g*t**2 + v0*np.sin(i)*t

vx = v0*np.cos(i)

vy = -g*t + v0*np.sin(i)
```

```python
plt.plot(x,y,"bx")
#axes = plt.gca()
#axes.set_aspect('equal','box')
plt.quiver(x,y,vx,vy,color='r',scale=500)
plt.show()
```
